```python
import psycopg2 as pg2
import random
import re

conn = pg2.connect(database='DvdRental', user='postgres',password='1400',port=5432,host='localhost')
cur = conn.cursor()


def menu():
    print('1.Create New User.')
    print('2.Show User.')
    print('3.Delete User.')
    print('4.Deposit money.')
    print('5.Withdraw Money.')
    print('6.Transfer MOney.')
    print('7.Show All Users.')
    print('8.Exit.')

def getting_info():
    firstname = input('Enter Your First Name: ')
    lastname = input('Enter Your Last Name: ')
    password = password_verify()
    id = id_generator()
    cash = input('Enter Your Initial Amount: ')
    cur.execute("INSERT INTO users(firstname,lastname,password,id,amount) VALUES(%s,%s,%s,%s,%s)",(firstname,lastname,password,id,cash))
    conn.commit()
    print('User Successfully Created\n')
        
def id_generator():
    id = random.randint(999,10000)
    cur.execute('SELECT * FROM users')
    data = cur.fetchall()
    for items in data:
        if items[0] != id:
            return id
        else:
            print('ID Already Exists\n')
            
def user_verify():
    user_verify = False
    while True:
        id = int(input('Enter Your Id: '))
        password = input('Enter Your Password Here: ')
        cur.execute('SELECT * FROM users')
        data = cur.fetchall()
        for items in data:
            if items[0] == id and items[3] == password:
                print('User Found\n')
                user_verify = False
                return items
        else:
            print('Invalid Credentials!\n')
            user_verify = True
            
            
def deposit():
    user = user_verify()
    amount = int(input('Enter Amount You Want To Deposit: '))
    updated_amount = user[4] + amount
    cur.execute(f'UPDATE users SET amount= {updated_amount} WHERE id = {user[0]}')
    conn.commit()
    print('Amount Updated Successfully\n')
            
                    
def withdraw():
    updated_flag = False
    invalid_amount_flag = True
    user = user_verify()
    while invalid_amount_flag:
        amount = int(input('Enter The Amount You Want To Withdraw: '))
        if user[4] >= amount:
            updated_amount = user[4] - amount
            cur.execute(f'UPDATE users SET amount = {updated_amount} WHERE id = {user[0]}')
            conn.commit()
            invalid_amount_flag = False
            updated_flag = True
        else:
            print('Invalid Amount!\n')
            invalid_amount_flag = True
    if updated_flag:
        print('Withdraw Successful!\n')
            
            
def show_user():
    user = user_verify()
    print(user)

    
def transfer():
    confirm_flag = False
    insufficient_flag = True
    giver = user_verify()
    reciever = id_verification()
    while insufficient_flag:
        amount = int(input('Enter The Amount You Want To Tranfer: \n'))
        if giver[4] >= amount:
            updated_amount_giver = giver[4] - amount
            cur.execute(f'UPDATE users SET amount = {updated_amount_giver} WHERE id = {giver[0]}')
            conn.commit()
            updated_amount_reciever = reciever[4] + amount
            cur.execute(f'UPDATE users SET amount = {updated_amount_reciever} WHERE id = {reciever[0]}')
            conn.commit()
            confirm_transfer = True
            insufficient_flag = False
            break
        else:
            print('Insufficient Funds\n')
            insufficient_flag = True
    if confirm_transfer:
        print('Transfer Successful!\n')
            
            
def id_verification():
    confirm_id = False
    while True:
        id = int(input('Enter recievers Id: '))
        cur.execute('SELECT * FROM users')
        data = cur.fetchall()
        for items in data:
            if items[0] == id:
                print('id matched')
                return items
        else:
            print('Invalid Id!\n')
            
            
def show_all():
    cur.execute('SELECT * FROM users')
    data = cur.fetchall()
    for items in data:
        print(items)


def delete_user():
    user = user_verify()
    cur.execute(f'DELETE FROM users WHERE id = {user[0]}')
    conn.commit() 
    print('User Deleted!\n')
    
    
def password_verify():
    pattern = re.compile(r'^(?=.*[A-Z])(?=.*[a-z])(?=.*[!@#$%^&*])(?=.*\d).+$')
    while True:
        password = input('Enter Strong Password: \n')
        if len(password) >= 8:
            if pattern.match(password):
                return password
            else:
                print("\nPassword should contain at least:\n1 special character\n1 number!\n1 uppercase letter\n"
                          "1 lowercase letter\n")
        else:
            print("\nPassword length less then 8 characters\n")
```
